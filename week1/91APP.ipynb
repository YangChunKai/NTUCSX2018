```python
def is_chinese(uchar):         
    if '\u4e00' <= uchar<='\u9fff':
        return True
    else:
        return False
```

```python
list(df['City'])
```

```python
import numpy as np
array_clean1 = np.array(df)
array_clean1
```

```python
drop_list = list() #找出需要清的index
ls_city = list(df['City'])
for i in range(len(ls_city)):
    try:
        if not (is_chinese(ls_city[i])):
#             print(str(i)+':'+(ls_city[i])+' will be dropped')
            drop_list.append(i)
            
    except:
#         print('error at '+ str(df['City'][i])+', i = '+str(i))
        drop_list.append(i)
        continue
array_clean1 = np.delete(array_clean1, drop_list, 0)
```

```python
df_clean1 = pd.DataFrame(array_clean1, columns = list(df))
df_clean1
```

```python
df_clean1_city=df_clean1.groupby(by='City')['Quantity'].sum()
df_clean1_city
```

```python
for i, city in enumerate(df_clean1['City']):
    if city =='桃園縣':
        df_clean1.set_value(i, 'City', '桃園市')
        
    elif city == '台北縣':        
        df_clean1.set_value(i, 'City', '新北市')
        
    elif city == '台東市':
         df_clean1.set_value(i, 'City', '台東縣')
            
    elif city == '屏東市':        
        df_clean1.set_value(i, 'City', '屏東縣')
        
    elif city == '台南縣':        
        df_clean1.set_value(i, 'City', '台南市')
df_clean1_city=df_clean1.groupby(by='City')['Quantity'].sum()
df_clean1_city
```
