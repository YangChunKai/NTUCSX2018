## 抓取[蘋果新聞](https://tw.appledaily.com/)關於大同公司的報導
### 將資料存到newslist中，包含
    1. 日期
    2. 標題
    3. 預覽內文
    4. 新聞連結

```python
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import requests
import re
from urllib.parse import quote
import time
import calendar
%matplotlib inline
```

```python
newslist=[]
for i in range(1,13):
    month = "%02d"%i
    date = str(calendar.monthrange(2017,i)[1])
    url="https://tw.appledaily.com/search/result?sort=time&searchType=all&dateStart=2017%2F"+month+"%2F01&dateEnd=2017%2F"+month+"%2F"+date+"&querystrS=大同公司"
    html = requests.get(url)
    time.sleep(0.2)
    soup = BeautifulSoup(html.text, 'html.parser')
    news=soup.select('.content')
    for i in range(len(news)):
        title=news[i].select('a')[0].text
        title=title.replace('\r','').replace('\n','').replace('\t','').replace(' ','').replace('    ','')
        newsdate=news[i].select('time')[0].text
        content=news[i].select('.ellipsis')[0].text
        link=news[i].select('a')[0]['href']
        ll=[]
        ll.append(newsdate)
        ll.append(title)
        ll.append(content)
        ll.append(link)
        newslist.append(ll)       
    
```

### 定義function finds('關鍵字')來搜尋包含關鍵字的新聞
### 印出包含關鍵字的新聞的時間、標題、新聞連結

```python
def finds(s):
    for i in range(len(newslist)):
        if s in newslist[i][2] or s in newslist[i][3]:
            print(newslist[i][0],newslist[i][1], newslist[i][3])
```
