```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import jieba
% matplotlib inline
```

```python
f = open('大同.txt','r', encoding ="utf-8")
demo_text = f.read()
```

```python
stopwords = ['我', '對', '之','於','註','係','民國','千元','資產','負債','年','價值','年度','臺','幣','大同','及','與','公司','公','司','為','月','有限公司','106','以','1','或','其他','金額','臺幣','1','2','3','4','5','000','認列','衡量','\n']
puncs = '！？｡＂＃＄$％＆＇()（）＊＋，_－／：；＜＝＞＠［＼］＾＿｀｛｜｝～｟｠｢｣､、〃》「」『』【】〔〕〖〗〘〙〚〛〜〝〞〟〰〾〿–—‘’‛“”„‟…‧﹏.。'
```

```python
seg_list = jieba.cut(demo_text, cut_all=False)  # 精确模式
print("Default Mode: " + "/ ".join(seg_list))
```

```python
seg_list = jieba.cut(demo_text, cut_all=False)
seg_list = list(seg_list)
wordDict = {}
for w in seg_list:
    if (w not in stopwords) and (w not in puncs):
        if w in wordDict:
            wordDict[w] = wordDict[w] + 1
        else:
            wordDict[w] = 1
print(wordDict)
```

```python
font = r"C:/Users/killingmachine/Desktop/Week_4/msjh.ttc"
wordcloud = WordCloud(font_path=font)
wordcloud.fit_words(wordDict)
plt.imshow(wordcloud)
```
